```python
import pandas as pd
import numpy as np
```

```python
# En primer lugar importamos nuestro documento.

basket=pd.read_csv("./input/basket_stats.csv")
```

```python
# Vamos a comprobar ahora los datos de nuestra base de datos.

basket
```

```python
# Vamos a ver ahora que columnas de datos tenemos.

basket.columns
```

```python
# Seleccionamos las columnas que vamos a utilizar a ver ahora que columnas de datos tenemos.

basket1=basket[['League', 'Season', 'Stage', 'Player', 'Team', 'GP', 'MIN', 'FGM',
       'FGA', '3PM', '3PA', 'FTM', 'FTA', 'TOV', 'PF', 'ORB', 'DRB', 'REB',
       'AST', 'STL', 'BLK', 'PTS']]
```

```python
# Vamos a renombrar algunas características para hacerlas más entendibles.

basket1.columns=['League', 'Season', 'Stage', 'Player', 'Team', 'GP', 'MIN', '2PM',
       '2PA', '3PM', '3PA', '1PM', '1PA', 'TOV', 'PF', 'ORB', 'DRB', 'REB',
       'AST', 'STL', 'BLK', 'PTS']
```

```python
# Vamos a seleccionar ahora unicamente la NBA, ya que nuestro cliente pertenece a dicha liga.

basket2=basket1[basket1["League"]=="NBA"]
```

```python
#Vamos a seleccionar sólo la fase de liga regular.

basket3=basket2[basket2["Stage"]=="Regular_Season"]
```

```python
# Seleccionamos ahora las dos últimas temporadas.

basket4=basket3[basket3["Season"]=="2018 - 2019"]
basket5=basket3[basket3["Season"]=="2019 - 2020"]
basket6=pd.concat([basket5, basket4], axis=0)
```

```python
# Seleccionamos ahora el equipo de nuestro cliente "Los Angeles Lakers" y al actual campeón de la NBA.
# El actual equipo de la NBA son los Toronto Raptors, los cuales ya utilizan ánalisis de datos para mejorar sis stats.

basket7=basket6[basket6["Team"]=="LAL"]
basket8=basket6[basket6["Team"]=="TOR"]
basket9=pd.concat([basket7, basket8], axis=0)
```

```python
# Vamos a comprobar ahora que no tengamos filas duplicadas.

basket9[basket9.duplicated()]==True
```

```python
# No tenemos filas duplicadas, por lo que vamos a comprobar ahora si tenemos valores nulos.

basket9.isnull().sum()
```

```python
# Vamos a utilizar la función regex para modificar los nombres de los equipos y hacerlos más entendibles.

basket9.Team=basket9.Team.str.replace(r" ?LAL ?", "Lakers")
basket9.Team=basket9.Team.str.replace(r" ?TOR ?", "Toronto Raptors")
```

```python
# Tampoco tenemos valores nulos, por lo que finalmente, está será la tabla que analizaremos.

basket9.reset_index()
```

```python
# Exportamos finalmente nuestra nueva tabla.

basket9.to_csv(r"./output/basket_analysis.csv", index = False)
```
